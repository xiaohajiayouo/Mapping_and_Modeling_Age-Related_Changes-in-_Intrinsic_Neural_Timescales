```python
import igraph as ig
from igraph import Graph
g = Graph.Erdos_Renyi(100, p = 0.16)
# print(g)
layout2 = g.layout("circle")
ig.plot(g,'', layout = layout2, edge_color="black", edge_width = 1, vertex_size = 15, vertex_frame = 5,\
        vertex_color = "black", edge_curved = 0.4, vertex_label="")
# ig.plot(g, layout = layout2, edge_color="black", edge_width = 1, vertex_size = 12, vertex_frame = 5,\
#         vertex_color = "#1b1bb3", edge_curved = 0.4, vertex_label="")
```

```python
import numpy as np
removing_node = np.random.choice(30, 3)
colors = ['#C0392B' if i not in removing_node else 'White' for i in range(30)]
print(colors)
ig.plot(g, layout = layout2, edge_color="black", edge_width = 3, vertex_size = 33,\
        vertex_color =colors, edge_curved = 0.4, vertex_label="")
```

```python
el = g.get_edgelist()
removing_conn = []
for i in removing_node:
   for e in el:
      if i in e:
        removing_conn.append(e)
print(removing_conn)        
g.delete_edges(removing_conn)
ig.plot(g, layout = layout2, edge_color="black", edge_width = 3, vertex_size = 33,\
        vertex_color = colors, edge_curved = 0.4, vertex_label="")
```

```python
el2 = g.get_edgelist()
print(len(el2))
removing_conn2 = np.random.choice(len(el2), int(len(el2)/3))
print(removing_conn2)
g.delete_edges(removing_conn2)
ig.plot(g,'elderly.png', layout = layout2, edge_color="black", edge_width = 4, vertex_size = 40,\
        vertex_color = colors, edge_curved = 0.4, vertex_label="")
```
